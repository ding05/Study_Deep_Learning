```
# Problem 4

# (a)

import matplotlib.pyplot as plt
import math
from decimal import Decimal

k = list(range(1, 51))
a = [0.4, 1, 0.01]
f = [[], [], []]
logf = [[], [], []]

for i in range(0, 3):
  w1 = -12
  w2 = 16
  for j in k:
    w1 = w1 - a[i] * (2.5 * w1 + 15)
    w2 = w2 - a[i] * (2 * w2 - 16)
    y = 1.25 * (w1 + 6) ** 2 + (w2 - 8) ** 2
    if y == 0:
      n = j + 7
      y += float(Decimal('1E-'+str(n)))
    # Add 1E-n to prevent turning to 0 starting the 23th iteration when a = 0.4.
    f[i].append(y)
    logf[i].append(math.log10(y))
  plt.plot(k, logf[i])

plt.legend((r'$a^{(k)}=\frac{1}{2.5}$', r'$a^{(k)}=1$', r'$a^{(k)}=0.01$'))
plt.ylabel(r'$\ln(f(w_1,w_2))$')
plt.xlabel(r'$k$')
plt.title("Fixed step size")
plt.show()
```

```
# (b)

import matplotlib.pyplot as plt
import math

k = list(range(1, 51))
a = [0.4, 1, 0.01]
n = [[], [], []]
logn = [[], [], []]

for i in range(0, 3):
  w1 = -12
  w2 = 16
  for j in k:
    w1 = w1 - a[i] * (2.5 * w1 + 15)
    w2 = w2 - a[i] * (2 * w2 - 16)
    y = math.sqrt(w1 ** 2 + w2 ** 2)
    n[i].append(y)
    logn[i].append(math.log10(y))
  plt.plot(k, logn[i])

plt.legend((r'$a^{(k)}=\frac{1}{2.5}$', r'$a^{(k)}=1$', r'$a^{(k)}=0.01$'))
plt.ylabel(r'$\ln(||\Delta f(w_1,w_2)||_2)$')
plt.xlabel(r'$k$')
plt.title("Fixed step size")
plt.show()
```
